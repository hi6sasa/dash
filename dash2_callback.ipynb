```python
import dash
import dash_html_components as html
import dash_core_components as dcc
import plotly.express as px
from dash.dependencies import Input,Output

core_style = {"width":"60%","margin":"5% auto"}

app = dash.Dash(__name__)

app.layout = html.Div(
    [
        #html
        html.H1("hello DAsh",style={"color":"red","textAlign":"center"}),
        
        #dropdown
        dcc.Dropdown(
            id = "my_dropdown",
            options = [
                {"label":"white","value":"white"},
                {"label":"yellow","value":"yellow"}
            ],
        value = "white",
        style = core_style,
        ),
        
        #graph
        dcc.Graph(
            figure = px.bar(x=[1,2,3,4,5],y=[1,2,3,4,5])
        )
    ],
    id = "all_components"
)

#callback
@app.callback(
    Output("all_components","style"),
    Input("my_dropdown","value"),
)

def update_background(selected_value):
    return{"backgroundColor":selected_value,"padding":"3%"}

if __name__ == "__main__":
    app.run_server(port=8070)
    #app.run_server(debug=True)
```
