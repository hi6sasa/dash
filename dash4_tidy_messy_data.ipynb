# 雑然messyデータは、人は理解しやすい。が機械は理解しづらい。機械は整然tidyデータがいいためmelt溶かす。
・各変数dimentionが１列で構成されている
・各観測measureが１行で構成されている
・観測単位が１つの表で構成されている

```python
import datetime
import pandas as pd

wide_df = pd.DataFrame(
    [
        ["TOKYO", 5.8, 5.7, 5.6],
        ["OSAKA", 6.8, 6.7, 6.4],
        ["NAGOYA", 5.1, 5.1, 5.0] 
    ],
    columns = ["location"] + pd.date_range("2020-12-01",periods=3).tolist(),
    #test
    #columns = ["location"] + ["20201"]+["20202","20223"],
)

wide_df
```

```python
tidy_df = wide_df.melt(
    #TOKYOは2020-12-01で5.8 2020-12-02で5.7 2020-12-03で5.6と解読してテーブルへ溶かしていくイメージ  
    id_vars = "location",var_name="datedesu",value_name = "Temprature"
    #test
    #id_vars = "20201",var_name="23",value_name = "Temprature"
)
tidy_df
```

```python
import plotly.express as px

#px.bar(tidy_df).show()
px.bar(tidy_df,x="datedesu",y="Temprature").show()
px.bar(tidy_df,x="datedesu",y="Temprature",color="location").show()
px.bar(tidy_df,x="datedesu",y="Temprature",color="location",barmode="group").show()
```
