```python
import numpy as np
import matplotlib.pyplot as plt
xs = np.linspace(0,10,100)
sins = np.sin(xs)
rand = np.random.rand(100)

#plt.plot(sins)
plt.scatter(xs,sins)
plt.scatter(xs,rand)
plt.show
```

```python
import plotly.graph_objects as go

fig = go.Figure(
    data=[
    go.Scatter(x=xs,y=sins,name="sin",mode="lines+markers"),
    go.Scatter(x=xs,y=rand,name="random"),
    ]
)

#another expression
# fig = go.Figure()
# fig.add_trace(go.Scatter(x=xs, y=sins, name="sin"))
# fig.add_trace(go.Scatter(x=xs, y=randoms, name="random"))

#fig.show()

#axes title
fig.update_xaxes(title="x")
fig.update_yaxes(title="y")

#axes range
fig.update_xaxes(range=(3,9))
fig.update_xaxes(rangeslider={"visible":True})

fig.update_yaxes(scaleanchor="x",scaleratio=2)

#figlayout config
fig.update_layout(title="title")
fig.update_layout(font={"family":"Meiryo","size":20})
fig.update_layout(showlegend=True)
fig.update_layout(width=800,height=600)
#fig.update_layout(xaxis_type="linear",yaxis_type="log")

#fig.update_layout(template="plotly_dark")
fig.update_layout(template="ggplot2")
#select theama
#['ggplot2', 'seaborn', 'simple_white', 'plotly',
#'plotly_white', 'plotly_dark', 'presentation', 'xgridoff',
#'ygridoff', 'gridon', 'none']
```

```python
#Y is 2
from plotly.subplots import make_subplots

fig = make_subplots(specs=[[{"secondary_y":True}]])
fig.add_trace(go.Scatter(x=xs,y=sins,name="sin"),secondary_y=False)
fig.add_trace(go.Scatter(x=xs,y=rand,name='random'),secondary_y=True)
fig.update_yaxes(title_text="y1",secondary_y=False)
fig.update_yaxes(title_text="y2",secondary_y=True)
fig.show()
```

```python
#several fig
from plotly.subplots import make_subplots

fig = make_subplots(rows=1, cols=2)
fig.add_trace(go.Scatter(x=xs, y=sins, name="sin"), row=1, col=1)
fig.add_trace(go.Scatter(x=xs, y=rand, name="random"), row=1, col=2)
fig.show()
```
